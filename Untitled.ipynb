```python
from konlpy.corpus import kolaw
```

```python
kolaw.fileids()
```

data수집

```python
t_data=kolaw.open('constitution.txt').read()
#하나의 문서로 구성되있다.
t_data
```

2. 전처리(형태소 분석기)

```python
from konlpy.tag import Okt
tw=Okt()
n_t_data=tw.nouns(t_data)
n_t_data
```

```python
tw.tagset
```

```python
type(n_t_data) #단어 집합들
```

 3. 시각화 작업을 위한 nltk.Text클래스 이용

```python
from nltk import Text #텍스트문서 분석 모듈
ck_data=Text(n_t_data,name='kolaw')
print(ck_data)
print(type(ck_data))
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumGothic')
ck_data.plot(15)
plt.show()
```

```python
ck_data.vocab()
```
